```python
import bruh
from matplotlib import pyplot as plta
```

```python
video = bruh.Video()
waited = 0
```

```python
while not video.frame_available():
        waited += 1
        print('\r  Frame not available (x{})'.format(waited), end='')
        # cv2.waitKey(30)
```

```python
if video.frame_available():
    frame = video.frame()
```
